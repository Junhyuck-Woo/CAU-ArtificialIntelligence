```python
from __future__ import print_function
import os
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data_path = ['C:\\Users\\ys\\Documents\\AI Project2\\data']
file_path = os.sep.join(data_path + ['Wine_Quality_Data.csv'])
data = pd.read_csv(file_path)
```

```python
# String to Number
from sklearn.preprocessing import LabelBinarizer

lb = LabelBinarizer()

for col in ['color']:
    data[col] = lb.fit_transform(data[col])
    
# Mute the sklearn warning
import warnings
warnings.filterwarnings('ignore', module='sklearn')

from sklearn.preprocessing import MinMaxScaler
msc = MinMaxScaler()

data = pd.DataFrame(msc.fit_transform(data), columns=data.columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from pandas.plotting import parallel_coordinates

plt.figure()
data.boxplot(by="color", figsize=(15, 10))
plt.show()

# plt.figure()
# sns.pairplot(data, hue = "color", size=3, markers=["o", "s", "D"])
# plt.show().plt.figure()
# dataset.boxplot(by="color", figsize=(15, 10))
# plt.show()
```

```python
# Data Processing
X_data = data[data.columns[:-1]]
y_data = data.color
```

```python
# Split as train data and test data
x_train, x_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2, shuffle=True)
```

```python
def accuracy(real, predict) :
    return sum(real == predict) / float(real.shape[0])
```

```python
# KNN, neighbors=3
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=3)

KNN = KNN.fit(x_train, y_train)
y_predict = KNN.predict(x_test)
```

```python
print(accuracy(y_test, y_predict))
```

```python
# KNN, neighbors=5
from sklearn.neighbors import KNeighborsClassifier

KNN = KNeighborsClassifier(n_neighbors=5)

KNN = KNN.fit(x_train, y_train)
y_predict = KNN.predict(x_test)
```

```python
print(accuracy(y_test, y_predict))
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier

DTC = DecisionTreeClassifier()

DTC = DTC.fit(x_train, y_train)
y_predict = DTC.predict(x_test)
```

```python
print(accuracy(y_test, y_predict))
```

```python
#Bernoulli Naive Bayes
from sklearn.naive_bayes import BernoulliNB

BNB = BernoulliNB(alpha=1.0)

BNB = BNB.fit(x_train, y_train)
y_predict = BNB.predict(x_test)
```

```python
print(accuracy(y_test, y_predict))
```

```python
#SVM
from sklearn.svm import LinearSVC

LinSVC = LinearSVC()

LinSVC = LinSVC.fit(x_train, y_train)
y_predict = LinSVC.predict(x_test)
```

```python
print(accuracy(y_test, y_predict))
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()

LR = LR.fit(x_train, y_train)
y_predict = LR.predict(x_test)
```

```python
print(accuracy(y_test, y_predict))
```
